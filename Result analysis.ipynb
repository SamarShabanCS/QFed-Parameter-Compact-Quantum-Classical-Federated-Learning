### epochs = 5

```python
# Load accuracy results for each method
qtFL_testacc_r1s_e5 = np.load('./results_FLfMnist5/qt_FL_test_acc_list_r_1c_5.dat', allow_pickle=True)
qtFL_trainacc_r1c1_e5 = np.load('./results_FLfMnist5/qt_FL_train_acc_list_r_1c_1.dat', allow_pickle=True)
qtFL_trainacc_r1c2_e5 = np.load('./results_FLfMnist5/qt_FL_train_acc_list_r_1c_2.dat', allow_pickle=True)
qtFL_trainacc_r1c3_e5 = np.load('./results_FLfMnist5/qt_FL_train_acc_list_r_1c_3.dat', allow_pickle=True)
qtFL_trainacc_r1c4_e5 = np.load('./results_FLfMnist5/qt_FL_train_acc_list_r_1c_4.dat', allow_pickle=True)
```

```python
# Generate x-axis (communication rounds)
rounds = np.arange(1, len(qtFL_trainacc_r1c1_e5) + 1)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(rounds, qtFL_testacc_r1s_e5, marker='o', linestyle='-', label='server', color='royalblue')
plt.plot(rounds, qtFL_trainacc_r1c1_e5, marker='s', linestyle='--', label='clients 1', color='forestgreen')
plt.plot(rounds, qtFL_trainacc_r1c2_e5, marker='^', linestyle='-.', label='clients 2', color='crimson')
plt.plot(rounds, qtFL_trainacc_r1c3_e5, marker='x', linestyle=':', label='clients 3', color='darkorange')
plt.plot(rounds, qtFL_trainacc_r1c4_e5, marker='D', linestyle='-', label='clients 4', color='mediumpurple')

# Plot settings
plt.title('flqt global and local models while rounds=1 & local epochs=5')
plt.xlabel('Communication Round')
plt.ylabel('Accuracy (%)')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
#plt.savefig('fl_accuracy_comparison.png', dpi=300)
plt.show()
```

```python
# Select the last round accuracy (or any specific round you prefer)
test_acc = qtFL_testacc_r1s_e5[-1] 
train_accs = [
    qtFL_trainacc_r1c1_e5[-1],
    qtFL_trainacc_r1c2_e5[-1],
    qtFL_trainacc_r1c3_e5[-1],
    qtFL_trainacc_r1c4_e5[-1]
]

# Bar labels
labels = ['Client 1', 'Client 2', 'Client 3', 'Client 4', 'server']

# Combine all into one list for plotting
all_accuracies_r1_e5 = train_accs + [test_acc]

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(labels, all_accuracies, color=['skyblue']*4 + ['navy'])
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.1f}%', ha='center', va='bottom')


plt.ylabel('Accuracy (%)')
plt.title(' one round Training Accuracies over 4 clients and server test accuracy (local epochs =5)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
#plt.savefig('qtfl_accuracy_barplot.png', dpi=300)
plt.show()
```
